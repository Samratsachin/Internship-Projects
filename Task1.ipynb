```python
# Population Data Visualization

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv'  # Replace with your file path
df = pd.read_csv(file_path, skiprows=4)
```

```python
# Select latest year
latest_year = '2023'
df_year = df[['Country Name', latest_year]].dropna()
```

```python
# Top 20 countries by population
top20 = df_year.sort_values(by=latest_year, ascending=False).head(20)
```

```python
# Bar Chart
plt.figure(figsize=(12,6))
plt.bar(top20['Country Name'], top20[latest_year])
plt.xticks(rotation=90)
plt.title(f'Population by Country in {latest_year} (Top 20 Countries)')
plt.ylabel('Population')
plt.xlabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8,6))
plt.hist(df_year[latest_year], bins=20, edgecolor='black')
plt.title(f'Distribution of Population Across Countries in {latest_year}')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
